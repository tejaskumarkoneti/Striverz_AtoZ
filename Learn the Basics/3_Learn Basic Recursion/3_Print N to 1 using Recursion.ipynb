```python
# print n to 1 backwards
def show(n):
    
    if n == 0:
        return
    print(n,end=" ")
    show(n-1)
    


show(3)
```

### Backtracking

```python
def ntoone(i,n):
    if i > n:
        return
    
    ntoone(i+1,n)
    print(i,end=" ")

num = int(input())
ntoone(1,num)
```
