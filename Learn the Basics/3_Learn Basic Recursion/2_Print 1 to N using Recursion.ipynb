## In function we can you use more than 1 parameters

```python
# print values from 1 to N

def oneton(i,n):
    
    if i > n:
        
        return
    
    print(i,end=" ")
    
    oneton(i+1,n)
    
num = int(input())

oneton(1,num)
```

## backtracking

```python
# print values from 1 to N

def oneton(i,n):
    
    if i < 1:
        
        return
    
    oneton(i-1,n)
    
    print(i,end=' ')
    
num = 5 
oneton(num,num)
```
