```python
n2 = 12345
temp_n = n2

rev = ""

# loop runs till n2 value is > than 0
while n2 > 0:
    
    last = n2 % 10 
    # we will get last digit of n2 
    # in each iteration 
    
    rev = rev + str(last)
    # we  will add the number from last digit value to rev
    
    n2 = n2 // 10
    # n2 will divide by 10 to igrore the last value of n2

print("The reverse of a number {} is {}".format(temp_n,rev))    
    
```
