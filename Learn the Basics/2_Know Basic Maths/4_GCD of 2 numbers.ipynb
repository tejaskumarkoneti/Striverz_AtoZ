### Better Approach

```python
N1 = 69
N2 = 25
print("GCD of {} and {} is :".format(N1,N2),end=" ")

for i in range(N2):
    
    if N2 == 0:
        break       
    elif N1 > N2:
        r1 = N1  % N2
        N1,N2 = N2,r1
    
    elif N2 > N1:
        r2 = N2 % N1
        N1,N2 = N2,r2

print(N1)
```
