```python
n = 12345
temp_n = n
# Variable to store reversed number
revNum = 0

# Loop until all digits are processed
while n > 0:
    # Get the last digit
    lastDigit = n % 10

    # Append it to the reversed number
    revNum = revNum * 10 + lastDigit

    # Remove the last digit from n
    n = n // 10
    
print(revNum)
```

```python
n3 = 1213

dup = n3 # store orginal n3 value

# Variable to store reversed number
revNum = 0

# Loop until all digits are processed
while n3 > 0:
    # Get the last digit
    lastDigit = n3 % 10

    # Append it to the reversed number
    revNum = revNum * 10 + lastDigit

    # Remove the last digit from n
    n3 = n3 // 10
    
if dup == revNum:  #dup  compare with revNum
    
    print("The reverse of {} is {} Palindrome Number".format(temp_n,revNum))
    
else:
    
    print("The reverse of {} is {} Not Palindrome".format(temp_n,revNum))
```
