### Brute Force  2/3/26

```python
N1 = 69
N2 = 25

for i in range(1,min(N1,N2)+1):
    
    if N1 % i == 0 and N2 % i == 0:
        
        gcd = i

print(gcd)
```

### Optimal Approach 2/3/26


```python
N1 = 69
N2 = 25

for i in range(min(N1,N2),0,-1):
    
    if N1 % i == 0 and N2 % i == 0:
        
        gcd = i

print(gcd)
```

### Better Approach

```python
N1 = 69
N2 = 25
print("GCD of {} and {} is :".format(N1,N2),end=" ")

for i in range(N2):
    
    if N2 == 0:
        break       
    elif N1 > N2:
        r1 = N1  % N2
        N1,N2 = N2,r1
    
    elif N2 > N1:
        r2 = N2 % N1
        N1,N2 = N2,r2

print(N1)
```
