```python
# print n to 1 backwards
def show(n):
    
    if n == 0:
        return
    print(n)
    show(n-1)

show(3)
```

```python
# Factorial of n 

def fact(n):
    
    if n == 0 or n == 1:
        
        return 1
    
    else:
        
        return n * fact(n-1)
    
fact(3)
```

```python
# sum of n natural numbers
def sum(n):
    
    if n == 0:
        
        return 0
    
    return n + sum(n-1)

sum(5)
```

```python
# sum of n natural numbers
def calc_sum(n):
    
    if n == 0:
        
        return 0
    
    return calc_sum(n-1) + n

sum = calc_sum(5)

print(sum)
```

```python
def print_name(count):
    
    if count == N:
        return
    print("Jeshu")
    print_name(count+1)

N = 3
print_name(0)
```

```python
class Solution:
    # Recursive function to print name count times
    def printName(self, name, count, N):
        # Base case: if count equals N, stop recursion
        if count == N:
            return

        # Print the name
        print(name)

        # Recursive call with incremented count
        self.printName(name, count + 1, N)

if __name__ == "__main__":
    sol = Solution()
    N = 5
    name = "Tej"

    sol.printName(name, 0, N)
```

## In function we can you use more than 1 parameters

```python
# print values from 1 to N

def oneton(i,n):
    
    if i > n:
        
        return
    
    print(i,end=" ")
    
    oneton(i+1,n)
    
num = int(input())

oneton(1,num)
```
