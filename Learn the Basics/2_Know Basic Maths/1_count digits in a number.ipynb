```python
# Brute Force Approach

n1 = 12345

dummy = n1
count = 0       # Initialise a counter to store the number of digits.

 
while n1 > 0:   # While N is greater than 0, execute the following:

    
    count +=1   # Increment the counter by 1

    n1 = n1//10 # remove the last digit of n1 in each iteration
    
print("The number {} has {} digits".format(dummy,count))
```

```python
# Optimal Approach

import math  

N = 12345

# Intialize the variable count and perform log base10 operations

count = int(math.log10(N)+1)

# log base 10 gives the number of digits in N
# We add 1 to ensure that count value is correct
# Even for that are powers of 10
# we convert to int to avoid floating numbers

print(count)
```
