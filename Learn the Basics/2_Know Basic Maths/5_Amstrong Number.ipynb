##### Amstrong Number is a number that is equal to the number of each digits in number which is rasied to power to total numbers digits in that number

```python
N = 153

dup_N = N

res = 0

while N > 0:
    last_digit = N % 10
    res = res + last_digit ** len(str(dup_N))
    N = N // 10

print("The result is {}".format(res))
if dup_N == res:
    
    print("Amstrong Number")

else:
    
    print("Not Amstrong Number")
```

```python
N = 153

dup_N = N

k = len(str(dup_N))

res = 0

while N > 0:
    last_digit = N % 10
    res = res + last_digit ** k
    N = N // 10

print("The result is {}".format(res))
if dup_N == res:
    
    print("Amstrong Number")

else:
    
    print("Not Amstrong Number")
```
