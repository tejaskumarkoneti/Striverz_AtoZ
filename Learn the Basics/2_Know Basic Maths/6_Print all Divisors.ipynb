```python
class Solution:
    # Function to find all divisors
    def getDivisors(self, N):
        # Create list to store divisors
        res = []

        # Loop from 1 to N
        for i in range(1, N + 1):
            # Check if i is a divisor of N
            if N % i == 0:
                # Add i to result
                res.append(i)

        # Return the list of divisors
        return res

# Create object of Solution class
sol = Solution()

# Input number
N = 36

# Call the function to get divisors
result = sol.getDivisors(N)

# Print the result
print("Divisors of", N, ":", *result)
```

### Optimal Approach

##### üåü If d divides(√∑) N, then N √∑ d also divides N

##### So for 36: if 3 is a divisor ‚Üí 36 √∑ 3 = 12 is also a divisor. Done!

##### Every pair mirrors itself around the square root (6):
###### 1  √ó  36   ‚Üê‚Üí   36  √ó  1
###### 2  √ó  18   ‚Üê‚Üí   18  √ó  2
###### 3  √ó  12   ‚Üê‚Üí   12  √ó  3
###### 4  √ó   9   ‚Üê‚Üí    9  √ó  4
###### 6 √ó 6   ‚Üê middle!
###### If you know one half of the pair, you automatically know the other!
###### When you find that 2 divides 36, you instantly know 18 (which is 36√∑2) also divides 36. You get two answers for the price of one check!

```python
import math

class Solution:
    # Function to get all divisors
    def getDivisors(self, N):
        # Create list to store divisors
        res = []

        # Loop from 1 to square root of N
        for i in range(1, int(math.isqrt(N)) + 1):
            # Check if i divides N
            if N % i == 0:
                # Add i to result
                res.append(i)

                # If N // i is not the same, add that too
                if i != N // i:
                    res.append(N // i)

        # Return the list of divisors
        return res

# Create object of Solution class
sol = Solution()

# Input number
N = 36

# Get divisors
result = sol.getDivisors(N)

# Print the result
print("Divisors of", N, ":", *result)
```
