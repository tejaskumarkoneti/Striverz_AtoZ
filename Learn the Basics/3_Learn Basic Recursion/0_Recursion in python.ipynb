```python
# print n to 1 backwards
def show(n):
    
    if n == 0:
        return
    print(n)
    show(n-1)

show(3)
```

```python
# print n to 1 backwards
def show(n):
    
    if n == 0:
        return
    print(n)
    show(n-1)
    print("END")

show(3)
```

```python
# Factorial of n 

def fact(n):
    
    if n == 0 or n == 1:
        
        return 1
    
    else:
        
        return n * fact(n-1)
    
fact(3)
```

```python
# sum of n natural numbers
def sum(n):
    
    if n == 0:
        
        return 0
    
    return n + sum(n-1)

sum(5)
```

```python
# sum of n natural numbers
def calc_sum(n):
    
    if n == 0:
        
        return 0
    
    return calc_sum(n-1) + n

sum = calc_sum(5)

print(sum)
```
