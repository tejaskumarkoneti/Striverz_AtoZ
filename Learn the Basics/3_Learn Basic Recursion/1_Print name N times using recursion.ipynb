```python
# print name N times using recursion

N = 3
def print_name(n):
    
    if n == 0:
        
        return
    
    print("Ashish")
    
    print_name(n-1)
    
print_name(3)
```

```python
def print_name(count):
    
    if count == N:
        return
    print("Jeshu")
    print_name(count+1)

N = 3
print_name(0)
```

```python
class Solution:
    # Recursive function to print name count times
    def printName(self, name, count, N):
        # Base case: if count equals N, stop recursion
        if count == N:
            return

        # Print the name
        print(name)

        # Recursive call with incremented count
        self.printName(name, count + 1, N)

if __name__ == "__main__":
    sol = Solution()
    N = 5
    name = "Tej"

    sol.printName(name, 0, N)
```
